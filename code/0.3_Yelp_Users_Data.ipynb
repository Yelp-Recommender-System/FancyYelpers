```python
## preps
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#specify your path
sys_path = "/Users/xli/Yelp/yelp_dataset"
```

```python
# Load three pre-processed datasets: reviews data, business data for selected candidate cities

## The restaurants in the five candidate cities
restaurants_ = pd.read_csv("business_data_subset.csv", index_col= None)
### reviews in the top ten candidate cities
ratings_ = pd.read_csv("reviews_data_subset.csv")
```

```python
## get a list of user ids that had reviews for our candidate cities
users_ids_subset = ratings_.user_id.unique().tolist()
```

```python
user_data_samp = []
count = 0
count_cap = 1e9

with open("{0}/yelp_academic_dataset_user.json".format(sys_path)) as f:
    for line in f:
        if json.loads(line)["user_id"] in users_ids_subset:
            user_data_samp.append(json.loads(line))
        count+= 1
        if count > count_cap:
            break     
user_data_samp = pd.DataFrame.from_dict(user_data_samp)
user_data_samp.to_csv("user_data_subset.csv", index=None)
```
